```python
def to_float(series):
    series = series.str.replace(',','.').str.replace('−','-').astype(float)
    return series
def to_float_coma(series):
    series =  series.str.replace(',','').astype(float)
    return series
def to_float_dot(series):
    series =  series.str.replace('.','').astype(float)
    return series
def line_prepender(filename, line):
    with open(filename, 'r+') as f:
        content = f.read()
        f.seek(0, 0)
        f.write(line.rstrip('\r\n') + '\n' + content)
def add_list(filename, lines):
    lines.reverse()
    for i in lines:
        line_prepender(filename, i)
def get_owner(string):
    if (string == 'The World Bank'):
        return 'wd:Q7164';
```

```python
def obs_lines(df: pd.DataFrame, i: int, j: str, struct, dim3: str, 
              Measure: str, value_type, unit_multiplier: str, col: str, unit_decimal: str):

    observation = ['\n']
    #print(df[j].loc[i])

    if (value_type == str):
        obs = 'c19:dataset-' + struct + '-oc' + j + 'r' + str(i) + ' a qb:Observation;\n'
        observation.append(obs)

        dataset = '    qb:dataSet  c19:dataset-' + struct + ';\n'
        observation.append(dataset)

        period = '    c19-dimension:refPeriod "' + str(pd.to_datetime(df['Date'].loc[i]))[:10] + '"^^xsd:date;\n'
        observation.append(period)

        id_country = '    c19-dimension:refArea wd:' + df[j].iloc[df.shape[0]-2] + ' ;\n'
        observation.append(id_country)

        if (type(df[str(j)].iloc[df.shape[0]-1]) == str):
            id_Subcountry = '    c19-dimension:refSubArea wd:' + df[j].iloc[df.shape[0]-1] + ' ;\n'
            observation.append(id_Subcountry)

        dim3 = '    c19-dimension:patient c19:' + dim3 + ' ;\n'
        observation.append(dim3)

        disease = '    c19-dimension:disease wd:Q84263196 ;\n'
        observation.append(disease)

        #year = 'sdmx-attribute:unitMeasure <http://dbpedia.org/resource/Year> ;\n'
        #observation.append(year)

        #dim = '    c19-dimension:ageInterval c19-interval:Age15plus ;\n'
        #observation.append(dim)

        #sex = '    sdmx-dimension:sex         sdmx-code:sex-F ;\n'
        #observation.append(sex)

        #unit_Measure = "    sdmx-attribute:unitMeasure  c19-unit:" + unit_msr + ";\n"
        #observation.append(unit_Measure)

        #unit_Mult = "    sdmx-attribute:unitMultiplier  " + unit_multiplier + ";\n"
        #observation.append(unit_Mult)

        #unit_Dec = "    sdmx-attribute:decimals  " + unit_decimal + ";\n"
        #observation.append(unit_Dec)

        #currency = '    sdmx-attribute:currency      ' + current + ' ;\n'
        #observation.append(currency)


        measure = '    c19-measure:' + Measure + '  ' + df[j].loc[i] + ' .\n'
        observation.append(measure)

    #if (value_type == np.int64 or value_type == np.float64):
    if(str(df[j].loc[i]) != 'nan'):
        obs = 'c19:dataset-' + struct + '-oc' + col + 'r' + str(i) + ' a qb:Observation;\n'
        observation.append(obs)

        dataset = '    qb:dataSet  c19:dataset-' + struct + ';\n'
        observation.append(dataset)

        period = '    c19-dimension:refPeriod "' + str(pd.to_datetime(df['Date'].loc[i]))[:10] + '"^^xsd:date;\n'
        observation.append(period)

        id_country = '    c19-dimension:refArea wd:' + df['ID'].iloc[i] + ' ;\n'
        observation.append(id_country)

        dim3 = '    c19-dimension:patient c19:' + dim3 + ' ;\n'
        observation.append(dim3)

        disease = '    c19-dimension:disease wd:Q84263196 ;\n'
        observation.append(disease)

        unit_Mult = "    sdmx-attribute:unitMultiplier  " + unit_multiplier + ";\n"
        observation.append(unit_Mult)

        unit_Dec = "    sdmx-attribute:decimals  " + unit_decimal + ";\n"
        observation.append(unit_Dec)

        measure = '    c19-measure:' + Measure + '  ' + str(df[j].loc[i]) + ' .\n'
        observation.append(measure)

    return observation
```
